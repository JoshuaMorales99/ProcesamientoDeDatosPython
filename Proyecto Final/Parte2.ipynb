### Transformaciones en el DataSet.

Para este apartado, realizaremos algunas consultas y graficos.

```python
# Mostrar los registros que tengan 'High Confidence'.
df[df['Confidence in estimate'] == "High Confidence"]
```

```python
# Almacenamos el listado de los 10 paises con mayor desperdicio de comida del DataSet.
Top10HighestFoodWaste = df.nlargest(10, 'combined figures (kg/capita/year)')

# Mostrar el listado de los top 10 paises con mayor desperdicio de comida del DataSet.
Top10HighestFoodWaste
```

```python
# Importamos la libreria para hacer graficos en Python.
import seaborn as sns

# Creamos el grafico y lo mostramos.
sns.barplot(x = Top10HighestFoodWaste['combined figures (kg/capita/year)'], y = Top10HighestFoodWaste['Country'])
plt.title("Top 10 Paises con mayor desperdicio alimenticio")
plt.show()
```

```python
# Almacenamos el listado de los 10 paises con menor desperdicio de comida del DataSet.
Top10LessFoodWaste = df.nsmallest(10, 'combined figures (kg/capita/year)')

# Mostrar el listado de los top 10 paises con menor desperdicio de comida del DataSet.
Top10LessFoodWaste
```

```python
# Creamos el grafico y lo mostramos.
sns.barplot(x = Top10LessFoodWaste['combined figures (kg/capita/year)'], y = Top10LessFoodWaste['Country'])
plt.title("Top 10 Paises con menor desperdicio alimenticio")
plt.show()
```
